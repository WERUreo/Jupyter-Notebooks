```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame,read_html
```

```python
base_url = 'http://espn.go.com/college-football/statistics/player/_/stat/rushing'
```

```python
next_page = '/sort/rushingYards/qualified/false/count/'
```

```python
rushing_stats = (pd.io.html.read_html(base_url))[0]
```

```python
for x in range(1,39):
    rushing_stats = rushing_stats.append((pd.io.html.read_html(base_url + next_page + `(x*40 + 1)`)),ignore_index=True)
    
```

```python
rushing_stats.fillna(method='pad',inplace=True)
```

```python
rushing_stats.columns = rushing_stats.iloc[0]
```

```python
rushing_stats = rushing_stats.reindex(rushing_stats.index.drop(0))
```

```python
rushing_stats = rushing_stats[rushing_stats.TD != 'TD']
```

```python
rushing_stats = rushing_stats[rushing_stats.TEAM != 'UNKWN']
```
